### Getting Started Querying your at-home Machine

To make sure you're set up for doing local AI/ML at home, this notebook will guide you to the first steps of running an LLM via vLLM and ensuring that your computers are appropriately connected.

#### Steps

1. Get your GPU machine all set up and running some sort of Linux-based operating system. I use [Ubuntu](https://ubuntu.com/tutorials/try-ubuntu-before-you-install?cp=close#1-before-you-start) and [Debian](https://www.debian.org/distrib/) at home, but you can choose whatever one you like! Set up your username, password and get the initial OS installed.

2. Check what the IP address of your machine is on your local network using the display (or looking at your router). Try pinging the IP address from your laptop. Debug any networking issues early! You might also want to make an SSH login so that it's easy to tinker with your machine without a monitor. 

3. Once everything is initially set up and running, set up your Python environment. I like to use [Conda for this](https://docs.conda.io/projects/conda/en/latest/user-guide/install/), but use whatever makes sense for you - many people like [uv](https://docs.astral.sh/uv/getting-started/installation/).

4. Then make sure you install these packages (either via pip or uv): torch, vllm.

5. Create a hugging face login (or use one you already have) and [install the HuggingFace CLI](https://huggingface.co/docs/huggingface_hub/main/en/guides/cli) which will help you download models and datasets. You will need to create a [User Access token](https://huggingface.co/docs/hub/security-tokens) and save it on your machine for later use.

6. Test your login by using: `huggingface-cli auth` (note, this command might now be hf instead of huggingface-cli).

7. Once that's running, try downloading a model: `huggingface-cli download meta-llama/Meta-Llama-3-8B-Instruct`

8. When the download completes, run it via vLLM! `vllm serve meta-llama/Meta-Llama-3-8B-Instruct`

9. Then, use your local IP address in the notebook below!


```python
from openai import OpenAI
```

```python
test_prompt = 'what day is today?'
```

```python
messages = [{
    'role': 'user',
    'content': test_prompt
}]
```

```python
messages 
```

```python
openai_api_key = "EMPTY"
openai_api_base = "http://192.168.1.81:8000/v1" # change this for your setup!
```

```python
client = OpenAI(
    api_key=openai_api_key,
    base_url=openai_api_base,
)
```

```python
completion = client.chat.completions.create(
    model="meta-llama/Meta-Llama-3-8B-Instruct",
    messages=messages
)
```

```python
print(completion.choices[0].message.content)
```
