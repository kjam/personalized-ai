## Parsing PDFs with Tables

I've been battling PDFs with tables for at least 10 years (you can read about it in my [first O'Reilly book]()) but don't bother because it's gotten much easier! 

You can learn more about [Docling](https://www.docling.ai/) workflows via their documentation.

Some other libraries to check out:

- [pypdf](https://pypi.org/project/pypdf/): Have a simple PDF and just need text, use this!
- [SpaCy Layout](https://github.com/explosion/spacy-layout): Use NLP tools directly in the library you use to extact!


```python
from docling.datamodel.base_models import InputFormat
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.pipeline_options import PdfPipelineOptions, TableFormerMode
import pandas as pd
```

```python
# Configure pipeline options for best table extraction quality
pipeline_options = PdfPipelineOptions(do_table_structure=True)
pipeline_options.table_structure_options.do_cell_matching = False  # Avoids merging rows/columns
pipeline_options.table_structure_options.mode = TableFormerMode.ACCURATE  # Most accurate model
```

```python
# Create the document converter with these options
doc_converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
    }
)
```

```python
conv_res = doc_converter.convert('../data/gebietseigene_pflanzen.pdf', page_range=(47, 52))
```

```python
df_list = []
for i, table in enumerate(conv_res.document.tables):
    df = table.export_to_dataframe(doc=conv_res.document)
    df_list.append(df)
```
